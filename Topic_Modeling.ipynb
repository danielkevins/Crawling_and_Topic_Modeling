**Case Folding**

```python
df_text['case_folding'] = df_text['full_text'].str.lower()
df_text.head()
```

**Remove Hastag**

```python
def remove_hastag(text):
  return re.sub(r'#\S+', '', text)

df_text['remove_hastag'] = df_text['case_folding'].apply(remove_hastag)
df_text.head()
```

**Remove URL**

```python
def remove_url(text):
  url_pattern = r"http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+"
  return re.sub(url_pattern, "", text)

df_text['remove_url'] = df_text['remove_hastag'].apply(remove_url)
df_text.head()
```

**Remove Punctuation**

```python
def remove_punctuation(text):
  return text.translate(str.maketrans('', '', string.punctuation))

df_text['remove_punctuation'] = df_text['remove_url'].apply(remove_punctuation)
df_text.head()
```

**Remove Number**

```python
def remove_number(text):
  return re.sub(r'\d+', '', text)

df_text['remove_number'] = df_text['remove_punctuation'].apply(remove_number)
df_text.head()
```

**Remove Whitespace**

```python
df_text['remove_whitespace'] = df_text['remove_number'].str.strip()
df_text.head()
```

```python
df_text['remove_whitespace'] = df_text['remove_whitespace'].str.replace(r'\s+', ' ')
df_text.head()
```

```python
df_text['cleaned'] = df_text['remove_whitespace']
```

**Stopwords**

```python
import nltk
nltk.download('stopwords')
stop_factory = StopWordRemoverFactory()
more_stopword = ["dengan", "ia","bahwa","oleh"]
sastrawi_stopwords = set(stop_factory.get_stop_words()+more_stopword)
stopword = stop_factory.create_stop_word_remover()
print(len(sastrawi_stopwords))
```

```python
nltk_stopwords = set(stopwords.words('indonesian'))
print(len(nltk_stopwords))
```

```python
custom_stopwords = {'ganjar', 'ganjarpranowo', 'pak', 'tingkat', 'tahun', 'yg', 'ga', 'gak', 'aja', 'banyak', 'kenal', 'salah', 'ke', 'di', 'nih', 'pt', 'amp', 'fani', 'sandiuno', 'kalo', 'tuh', 'nya', 'tbk', 'sigid', 'sih', 'mnang', 'gama', 'gtu', 'mahfud', 'varia', 'org', 'semangat', 'pimpin', 'sig', 'guru', 'mandeh', 'imbang', 'kayak', 'ken', 'vub'}
words_to_exclude = {'kasus','jumlahnya', 'jumlah', 'berjumlah', 'masalah', 'akhir', 'berakhir', 'berakhirnya', 'meangakhiri', 'terakhir',
                    'diakhiri', 'masalahnya', 'bulan', 'awal', 'hari', 'gunakan', 'menggunakan', 'siap', 'mulai', 'dimulai',
                    'dimulainya', 'memulai', 'cukup','asal', 'banyak', 'sebanyak', 'terbanyak', 'mengatakan',
                    'sedikit', 'termasuk', 'naik', 'baru', 'jangan', 'tambah'}
indonesian_stopwords = nltk_stopwords | sastrawi_stopwords | custom_stopwords
indonesian_stopwords -= words_to_exclude
print(len(indonesian_stopwords))
```

```python
def remove_stopwords(tokens):
    filtered_tokens = [token for token in tokens if token not in indonesian_stopwords]
    return filtered_tokens
```

**TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd


df_clean['tf_idf'] = df_clean['stemmed_text'].apply(lambda x: ' '.join(x))

# Inisialisasi TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # Batasi hingga 1000 fitur

# Fit dan transform teks yang sudah di-clean
X_tfidf = tfidf_vectorizer.fit_transform(df_clean['tf_idf'])

# Ambil nama fitur (kata-kata unik yang dipilih oleh TF-IDF)
feature_names = tfidf_vectorizer.get_feature_names_out()

# Konversi matriks TF-IDF ke DataFrame
tfidf_df = pd.DataFrame(X_tfidf.toarray(), columns=feature_names)

# Tampilkan beberapa baris pertama dari matriks TF-IDF
print(tfidf_df.head())
```

```python
# Check the TF-IDF weights for a specific document (e.g., first document)
document_index = 0  # You can change this index to inspect other documents
tfidf_weights = pd.DataFrame({
    'Feature': feature_names,
    'TF-IDF Weight': X_tfidf[document_index].toarray().flatten()
})

# Sort by the highest TF-IDF weight to see the most important features (words)
tfidf_weights_sorted = tfidf_weights.sort_values(by='TF-IDF Weight', ascending=False)

# Display the top features with highest weights for the first document
tfidf_weights_sorted.head(10)
```

```python
# Get the feature names (words) from the TF-IDF vectorizer
feature_names = tfidf_vectorizer.get_feature_names_out()

# Create a dictionary of terms and their corresponding TF-IDF scores
# Calculate the sum of TF-IDF scores for each term across all documents
tfidf_scores = {term: X_tfidf[:, idx].sum() for idx, term in enumerate(feature_names)}

# Generate a word cloud based on TF-IDF scores
wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(tfidf_scores)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from sklearn.decomposition import PCA

# Reduksi dimensi menggunakan PCA
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(X_tfidf.toarray())

# Visualisasikan hasil PCA
plt.figure(figsize=(10, 8))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], alpha=0.8, c='blue', edgecolors='k')
plt.title('PCA dari TF-IDF Vektor')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

```python
import gensim
from gensim import corpora
from gensim.corpora import Dictionary
from gensim.models import LdaModel
from gensim.models.coherencemodel import CoherenceModel
```

```python
tokenized_tolist = df_clean['stemmed_text'].tolist()
```

```python
# membuat dictionary
dictionary = Dictionary(tokenized_tolist)
```

```python
dictionary.filter_extremes(no_below=2, no_above=0.8)
```

```python
# print dictionary kata:id sampai 100 item
print(list(dictionary.token2id.items())[:100])
print('=================')
print('jumlah total dictionary: ' + str(len(dictionary)))
```

**Membuat Corpus**

```python
# make corpus from dictionary with bag-of-words
corpus = [dictionary.doc2bow(doc) for doc in tokenized_tolist]
```

```python
# human readable corpus
human_readable_corpus = []
for document in corpus:
    text_document = [(dictionary[id], freq) for id, freq in document]
    human_readable_corpus.append(text_document)

# Print the human-readable corpus
for document in human_readable_corpus[:5]:
    print(document)
```

**Coherence Score**

```python
coherence_scores = []
min_topics = 1
max_topics = 5
topic_range = range(min_topics, max_topics + 1)  # Specify the range of topic numbers to explore

for num_topics in topic_range:
    lda_model = LdaModel(corpus=corpus, num_topics=num_topics, id2word=dictionary)
    coherence_model = CoherenceModel(model=lda_model, texts=tokenized_tolist, dictionary=dictionary, coherence='c_v')
    coherence_score = coherence_model.get_coherence()
    coherence_scores.append(coherence_score)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(topic_range, coherence_scores)
plt.xlabel("Number of Topics")
plt.ylabel("Coherence Score")
plt.title("Score Coherence Setiap Topik")
plt.show()
```

```python
for i, num_topics in enumerate(range(min_topics, max_topics+1)):
    coherence_score = coherence_scores[i]
    print("Num Topics =", num_topics, " || Coherence Score =", coherence_score)
```

**Model LDA**

```python
best_num_topics = 5
chunksize = 2000
passes = 40
iterations = 1000
random_state = 42 # seed so will not re-generate or randomized for the next run

best_lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=best_num_topics,
                                 chunksize=chunksize, passes=passes, iterations=iterations,
                                 random_state=random_state)
```
